## Let's plot the data.

First let's define a function to save the figures.

```python
# Write a function to save the figures
PROJECT_ROOT_DIR = "images"

def save_fig(fig_id, tight_layout=True):
    path = os.path.join(PROJECT_ROOT_DIR, fig_id + ".png")
    print("Saving figure", fig_id)
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format='png', dpi=300)
```

**Now let's start with plotting the figures with non-categorical values.** Only ***Time Spent in Academic(hrs/day)***, ***Years of Internet Use***, ***Total Internet Usage(hrs/day)***. So let's start with plotting ***Time Spent in Academic(hrs/day)***.

But first let's write a function to plot the numerical data.

```python
def numerical_data_plot(dataframe, hist_alpha=0.6, color='crimson', kde_mul=1000,
                        title='Image Title', xlabel='X Label', ylabel='Y Label'):
    
    plt.figure(figsize=(15, 6))
    sns.set(font_scale=1.5)
    
    plt.subplot(121)
    
    count, bin_edges = np.histogram(dataframe)
    dataframe.plot(kind='hist', alpha=hist_alpha,
                   xticks=bin_edges, color=color)
    
    # Let's add a KDE plot
    mn, mx = plt.xlim()
    plt.xlim(mn, mx)
    kde_x = np.linspace(mn, mx, 300)
    kde = st.gaussian_kde(dataframe)
    plt.plot(kde_x, kde.pdf(kde_x) * kde_mul, 'k--', color=color)
    
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    
    plt.subplot(122)
    red_circle = dict(markerfacecolor='r', marker='o')
    dataframe.plot(kind='box', color=color, flierprops=red_circle)
```

```python
numerical_data_plot(college_df['Time Spent in Academic(hrs/day)'],
                    hist_alpha=0.6, color='darkslateblue', kde_mul=200,
                    title='Average time Spent in academic studies in a day(hours)',
                    xlabel='Time(hours)', ylabel='Number of Students')

plt.show()
```

Now that we've plotted the non-categorical features, we can finally plot the categorical values. Let's use histogram to plot these categorical values.

Let's first define a function to plot the histograms.

```python
def categorical_bar_plot(dataframe, rot=0, alpha=0.80, color = ['steelblue', 'crimson'],
                         title='Distribution', xlabel = 'Column Name', ylabel='Number of Students'):
    
    dataframe.value_counts().plot(kind='bar', rot=rot, alpha=alpha, color=color)
    
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
```

Now let's try scatter plot on the dataset. But first **let's define a function so that we can easily perform this operation.**

```python
def categorical_scatter_plot(dataframe, x_column, y_column, title,
                             y_label, x_label = 'Number of Students'):
    
    plt.figure(figsize=(10, 6))
    sns.set(font_scale=1.5)

    plt.plot(dataframe[x_column].loc[dataframe[y_column] == 'Good'].index,
             dataframe[x_column].loc[dataframe[y_column] == 'Good'],
             'bo', label = 'Good')

    plt.plot(dataframe[x_column].loc[dataframe[y_column] == 'Satisfactory'].index,
             dataframe[x_column].loc[dataframe[y_column] == 'Satisfactory'],
             'yo', label = 'Satisfactory')

    plt.plot(dataframe[x_column].loc[dataframe[y_column] == 'Average'].index,
             dataframe[x_column].loc[dataframe[y_column] == 'Average'],
             'go', label = 'Average')

    plt.plot(dataframe[x_column].loc[dataframe[y_column] == 'Not Satisfactory'].index,
             dataframe[x_column].loc[dataframe[y_column] == 'Not Satisfactory'],
             'ro', label = 'Not Satisfactory')
    
    plt.title(title, fontweight='bold')
    plt.xlabel(x_label, fontweight='bold')
    plt.ylabel(y_label, fontweight='bold')
    plt.legend(loc='upper right', fontsize=14)
    
    plt.tight_layout()
```

Now we'll check the distribution of categorical values against the `'Academic Performance'` of the attribute. But first let's write some functions that are necessary.

```python
def init_dictionary(dictionary, labels):
    for label in labels:
        dictionary[label] = []
        
def append_to_dict(dictionary, indexes, values):
    x = 0
    for index in indexes:
        dictionary[index].append(values[x])
        x += 1
        
def furnish_the_lists(labels, indexes, values):
    list_dif = [i for i in labels + indexes if i not in labels or i not in indexes]
    
    indexes.extend(list_dif)
    for i in range(len(list_dif)):
        values.append(0)
        
def append_dataframe_to_dict(dataframe, column_name, labels, dictionary):
    values = dataframe[column_name].value_counts().tolist()
    indexes = dataframe[column_name].value_counts().index.tolist()
    furnish_the_lists(labels, indexes, values)
    append_to_dict(dictionary, indexes, values)

    return dictionary
```

```python
def cat_vs_cat_bar_plot(dataframe, column_name, column_cat_list):
    good_result_df = dataframe.loc[dataframe['Academic Performance'] == 'Good']
    satisfactory_result_df = dataframe.loc[dataframe['Academic Performance'] == 'Satisfactory']
    average_result_df = dataframe.loc[dataframe['Academic Performance'] == 'Average']
    unsatisfactory_result_df = dataframe.loc[dataframe['Academic Performance'] == 'Not Satisfactory']

    labels = column_cat_list
    dictionary = {}

    init_dictionary(dictionary, labels)
    
    dictionary = append_dataframe_to_dict(good_result_df, column_name, labels, dictionary)
    dictionary = append_dataframe_to_dict(satisfactory_result_df, column_name, labels, dictionary)
    dictionary = append_dataframe_to_dict(average_result_df, column_name, labels, dictionary)
    dictionary = append_dataframe_to_dict(unsatisfactory_result_df, column_name, labels, dictionary)
    
    return dictionary
```

```python
def autolabel(rects):
    
    total_height = 0
    
    for rect in rects:
        total_height += rect.get_height()
        
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format("{:.2f}".format((height/total_height)*100)) + '%',
                    xy = (rect.get_x() + rect.get_width()/2, height),
                    xytext = (0, 3), # 3 points vertical offset
                    textcoords = "offset points",
                    ha = 'center', va = 'bottom')
```

```python
# Styling:

cust_palt = [
    '#111d5e', '#c70039', '#f37121', '#ffbd69', '#ffc93c'
]
```

```python
def ctn_freq(dataframe, cols, xax, hue = None, rows = 3, columns = 1):
    
    ''' A function for displaying numerical data frequency vs age and condition '''
    
    fig, axes = plt.subplots(rows, columns, figsize=(16, 12), sharex=True)
    axes = axes.flatten()

    for i, j in zip(dataframe[cols].columns, axes):
        sns.pointplot(x = xax,
                      y = i,
                      data = dataframe,
                      palette = cust_palt[:4],
                      hue = hue,
                      ax = j, ci = False)      
        j.set_title(f'{str(i).capitalize()} vs. Time Spent in Academic(hrs/week)')

        
        plt.tight_layout()
```

```python
ctn_freq(college_df,
         ['Total Internet Usage(hrs/day)', 'Years of Internet Use'],
         'Time Spent in Academic(hrs/day)', hue='Academic Performance',rows=2, columns=1)
```
