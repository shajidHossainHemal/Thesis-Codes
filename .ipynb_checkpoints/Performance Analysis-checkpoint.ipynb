```python
import numpy as np
import pandas as pd
import os
import random
import scipy.stats as st

random.seed(42)
```

```python
%matplotlib inline

import matplotlib as mlt
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
```

```python
# Write a function to save the figures
PROJECT_ROOT_DIR = "."
DATASET_ID = "Combined"
IMAGES_PATH = os.path.join(PROJECT_ROOT_DIR, "Figures", DATASET_ID)
os.makedirs(IMAGES_PATH, exist_ok = True)

def save_fig(fig_id, tight_layout=True, fig_extension="png", resolution=300):
    path = os.path.join(IMAGES_PATH, fig_id + "." + fig_extension)
    print("Saving figure", fig_id)
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format=fig_extension, dpi=resolution)
```

```python
sns.set(font_scale=1.2)
sns.set_style("whitegrid", {'axes.grid' : False})
```

```python
training_acc = [91.82, 65.44, 98.57, 65.64]
testin_acc = [68.08, 66.67, 66.67, 56.67]
precision = [67, 66, 66, 59]
recall = [68, 67, 67, 57]
f1_score = [67, 64, 64, 55]
```

```python
training_acc_KFold = [62.39, 62.16, 65.44, 62.99] 
test_acc_KFold = [51.9, 62.4, 64.3, 54.8]
```

```python
training_acc_GS = [62.37, 65.44, 93.66, 65.03]
test_acc_GS = [59.05, 67.62, 69.52, 57.62]
```

```python
training_acc_RS = [60.74, 66.05, 83.64, 64.21]
test_acc_RS = [61.43, 67.62, 71.43, 55.71]
```

```python
training_acc_BS = [60.51, 65.85, 83.84, 64.01]
test_acc_BS = [60.74, 69.05, 85.24, 56.67]
```

```python
labels = ['Decision Tree', 'Logistic Regression', 'Random Forest', 'Naive Bayes']
```

```python
font = {'family': 'times new roman',
        'color':  'black',
        'weight': 'bold',
        'size': 24,
        }

font2 = {'family': 'times new roman',
        'color':  'black',
        'weight': 'normal',
        'size': 22,
        }

width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(8, 6.2))
fig.subplots_adjust(top=0.5, bottom=0.1, hspace=0.5, wspace=0.2)

ax.bar(labels, training_acc, width, color=['steelblue', 'sandybrown', 'seagreen', 'firebrick'])


ax.set_ylabel('Training accuracy', fontdict=font)
ax.set_xlabel('Classification models', fontdict=font)
ax.set_xticklabels(labels, rotation = 20, fontdict=font2)
ax.set_yticks(list(filter(lambda x : x >= 0 and x<=100, ax.get_yticks().tolist())))
ax.set_yticklabels(labels=[int(x) for x in ax.get_yticks().tolist()], fontdict=font2)
# ax.legend(prop={'family': 'times new roman', 'size': 17.5}).set_title(title='Academic performance',
#                                                                     prop={'family': 'times new roman', 'size': 17.5})


save_fig('Training_Accuracy_of_Classification_Models')
plt.show()
```

```python
font = {'family': 'times new roman',
        'color':  'black',
        'weight': 'bold',
        'size': 24,
        }

font2 = {'family': 'times new roman',
        'color':  'black',
        'weight': 'normal',
        'size': 22,
        }

width = 0.40  # the width of the bars

fig, ax = plt.subplots(figsize=(8, 6.2))
fig.subplots_adjust(top=0.5, bottom=0.1, hspace=0.5, wspace=0.2)

x = np.arange(len(labels))

# ax.bar(labels, training_acc_GS, width, color=['steelblue', 'crimson', 'seagreen', 'salmon', 'royalblue', 'violet'])
ax.bar(x - width/2, training_acc_GS, width/2, label='Grid Search')
ax.bar(x, training_acc_RS, width/2, label='Randomized Search')
ax.bar(x + width/2, training_acc_BS, width/2, label='Bayesian Search')

ax.set_ylabel('Optimization training accuracy', fontdict=font)
ax.set_xlabel('Classification models', fontdict=font)
ax.set_xticks(x - width/2 + 0.1)
ax.set_xticklabels(labels, rotation = 20, fontdict=font2)
ax.set_yticks(list(filter(lambda x : x >= 0 and x<=100, ax.get_yticks().tolist())))
ax.set_yticklabels(labels=[int(x) for x in ax.get_yticks().tolist()], fontdict=font2)
ax.legend(prop={'family': 'times new roman', 'size': 15}).set_title(title='Optimization methods',
                                                                    prop={'family': 'times new roman', 'size': 15})


save_fig('Optimization_training_accuracy_of_Classification_Models')
plt.show()
```

```python
font = {'family': 'times new roman',
        'color':  'black',
        'weight': 'bold',
        'size': 24,
        }

font2 = {'family': 'times new roman',
        'color':  'black',
        'weight': 'normal',
        'size': 22,
        }

width = 0.40  # the width of the bars

fig, ax = plt.subplots(figsize=(8, 6.2))
fig.subplots_adjust(top=0.5, bottom=0.1, hspace=0.5, wspace=0.2)

x = np.arange(len(labels))

# ax.bar(labels, training_acc_GS, width, color=['steelblue', 'crimson', 'seagreen', 'salmon', 'royalblue', 'violet'])
ax.bar(x - width/2, test_acc_GS, width/2, label='Grid Search', color='steelblue')
ax.bar(x, test_acc_RS, width/2, label='Randomized Search', color='crimson')
ax.bar(x + width/2, test_acc_BS, width/2, label='Bayesian Search', color='seagreen')

ax.set_ylabel('Optimization testing accuracy', fontdict=font)
ax.set_xlabel('Classification models', fontdict=font)
ax.set_xticks(x - width/2 + 0.1)
ax.set_xticklabels(labels, rotation = 20, fontdict=font2)
ax.set_yticks(list(filter(lambda x : x >= 0 and x<=100, ax.get_yticks().tolist())))
ax.set_yticklabels(labels=[int(x) for x in ax.get_yticks().tolist()], fontdict=font2)
ax.legend(loc='upper left', prop={'family': 'times new roman', 'size': 15}).set_title(title='Optimization methods',
                                                                    prop={'family': 'times new roman', 'size': 15})


save_fig('Optimization_testing_accuracy_of_Classification_Models')
plt.show()
```
