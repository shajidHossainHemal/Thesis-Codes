## Plot the data

Now we can plot the data. Let's write a couple of functions so that we easily plot the data. 

**This function saves the figures.**

```python
# Write a function to save the figures
PROJECT_ROOT_DIR = "images"

def save_fig(fig_id, tight_layout=True):
    path = os.path.join(PROJECT_ROOT_DIR, fig_id + ".png")
    print("Saving figure", fig_id)
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format='png', dpi=300)
```

**This function plots histogram and box plot of the given non-categorical data.**

```python
def numerical_data_plot(dataframe, hist_alpha=0.6, color='crimson', kde_mul=1000,
                        title='Image Title', xlabel='X Label', ylabel='Y Label'):
    
    plt.figure(figsize=(15, 6))
    sns.set(font_scale=1.5)
    
    plt.subplot(121)
    
    count, bin_edges = np.histogram(dataframe)
    dataframe.plot(kind='hist', alpha=hist_alpha,
                   xticks=bin_edges, color=color)
    
    # Let's add a KDE plot
    mn, mx = plt.xlim()
    plt.xlim(mn, mx)
    kde_x = np.linspace(mn, mx, 300)
    kde = st.gaussian_kde(dataframe)
    plt.plot(kde_x, kde.pdf(kde_x) * kde_mul, 'k--', color=color)
    
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    
    plt.subplot(122)
    red_circle = dict(markerfacecolor='r', marker='o')
    dataframe.plot(kind='box', color=color, flierprops=red_circle)
```

**This function plots histograms of the given categorical data.**

```python
def categorical_bar_plot(dataframe, rot=0, alpha=0.80, color = ['steelblue', 'crimson'],
                         title='Distribution', xlabel = 'Column Name', ylabel='Number of Students'):
    
    dataframe.value_counts().plot(kind='bar', rot=rot, alpha=alpha, color=color)
    
    plt.title(title, fontweight='bold')
    plt.xlabel(xlabel, fontweight='bold')
    plt.ylabel(ylabel, fontweight='bold')
```

**let's define a function to create scatter plots of the numerical values and check the distribution of the attribute values against the target column, `Class Performance`**

```python
def categorical_scatter_plot(dataframe, x_column, y_column, title,
                             y_label, x_label = 'Number of Students'):
    
    plt.figure(figsize=(10, 6))
    sns.set(font_scale=1.5)

    plt.plot(dataframe[x_column].loc[dataframe[y_column] == 'Good'].index,
             dataframe[x_column].loc[dataframe[y_column] == 'Good'],
             'bo', label = 'Good')

    plt.plot(dataframe[x_column].loc[dataframe[y_column] == 'Satisfactory'].index,
             dataframe[x_column].loc[dataframe[y_column] == 'Satisfactory'],
             'yo', label = 'Satisfactory')

    plt.plot(dataframe[x_column].loc[dataframe[y_column] == 'Average'].index,
             dataframe[x_column].loc[dataframe[y_column] == 'Average'],
             'go', label = 'Average')

    plt.plot(dataframe[x_column].loc[dataframe[y_column] == 'Not Satisfactory'].index,
             dataframe[x_column].loc[dataframe[y_column] == 'Not Satisfactory'],
             'ro', label = 'Not Satisfactory')
    
    plt.title(title, fontweight='bold')
    plt.xlabel(x_label, fontweight='bold')
    plt.ylabel(y_label, fontweight='bold')
    plt.legend(loc='lower right', fontsize=14)
```

**This function plot histograms of the categorical values against the `'Class Performance'` column.** 

These are helper functions.

```python
def init_dictionary(dictionary, labels):
    for label in labels:
        dictionary[label] = []
        
def append_to_dict(dictionary, indexes, values):
    x = 0
    for index in indexes:
        dictionary[index].append(values[x])
        x += 1
        
def furnish_the_lists(labels, indexes, values):
    list_dif = [i for i in labels + indexes if i not in labels or i not in indexes]
    
    indexes.extend(list_dif)
    for i in range(len(list_dif)):
        values.append(0)
        
def append_dataframe_to_dict(dataframe, column_name, labels, dictionary):
    values = dataframe[column_name].value_counts().tolist()
    indexes = dataframe[column_name].value_counts().index.tolist()
    furnish_the_lists(labels, indexes, values)
    append_to_dict(dictionary, indexes, values)

    return dictionary
```

This is the main function.

```python
def cat_vs_cat_bar_plot(dataframe, column_name, column_cat_list):
    good_result_df = dataframe.loc[dataframe['Class Performance'] == 'Good']
    satisfactory_result_df = dataframe.loc[dataframe['Class Performance'] == 'Satisfactory']
    average_result_df = dataframe.loc[dataframe['Class Performance'] == 'Average']
    unsatisfactory_result_df = dataframe.loc[dataframe['Class Performance'] == 'Not Satisfactory']

    labels = column_cat_list
    dictionary = {}

    init_dictionary(dictionary, labels)
    
    dictionary = append_dataframe_to_dict(good_result_df, column_name, labels, dictionary)
    dictionary = append_dataframe_to_dict(satisfactory_result_df, column_name, labels, dictionary)
    dictionary = append_dataframe_to_dict(average_result_df, column_name, labels, dictionary)
    dictionary = append_dataframe_to_dict(unsatisfactory_result_df, column_name, labels, dictionary)
    
    return dictionary
```

**This function add value counts on top of each bar in the histogram.**

```python
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy = (rect.get_x() + rect.get_width()/2, height),
                    xytext = (0, 3), # 3 points vertical offset
                    textcoords = "offset points",
                    ha = 'center', va = 'bottom')
```

### Inspecting Age Closer

Let's define a function to make this process easier. 

```python
# For Styling:
cust_palt = [
    '#111d5e', '#c70039', '#f37121', '#ffbd69', '#ffc93c'
]
```

```python
def ctn_freq(dataframe, cols, xax, hue = None, rows = 3, columns = 1):
    
    ''' A function for displaying numerical data frequency vs age and condition '''
    
    fig, axes = plt.subplots(rows, columns, figsize=(16, 12), sharex=True)
    axes = axes.flatten()

    for i, j in zip(dataframe[cols].columns, axes):
        sns.pointplot(x = xax,
                      y = i,
                      data = dataframe,
                      palette = cust_palt[:4],
                      hue = hue,
                      ax = j, ci = False)      
        j.set_title(f'{str(i).capitalize()} vs. Age')

        
        plt.tight_layout()
```

**Now let's inspect the columns `'Total Internet Usage(hrs/week)'`, `'Years of Internet Use'`, `'Time Spent in Academic(hrs/week)'` against the column `'Age'` and also segment the distribution by the target `'Class Performance'`.**

```python
ctn_freq(college_df,
         ['Total Internet Usage(hrs/week)', 'Time Spent in Academic(hrs/week)', 'Years of Internet Use'],
         'Age', hue='Class Performance',rows=3, columns=1)
```

# Start Predicting the Models

```python
college = college_df.drop(["Age", "Class Performance"], axis = 1)
college_labels = college_df["Class Performance"].copy()
```

```python
college_labels.head()
```

```python
from sklearn.preprocessing import LabelBinarizer

encoder = LabelBinarizer()
pop_website = college["Popular Website"]
popular_website_1hot = encoder.fit_transform(pop_website)
popular_website_1hot
```

```python
from sklearn.base import BaseEstimator, TransformerMixin

class DataFrameSelector(BaseEstimator, TransformerMixin):
    def __init__(self, attribute_names):
        self.attribute_names = attribute_names
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        return X[self.attribute_names].values
```

```python
college.info()
```

```python
college_cat = college.drop(["Total Internet Usage(hrs/week)",
                               "Time Spent in Academic(hrs/week)", "Years of Internet Use"], axis = 1)
college_cat.head()
```

```python
college_num = college[["Total Internet Usage(hrs/week)", "Time Spent in Academic(hrs/week)",
                          "Years of Internet Use"]].copy()
college_num.head()
```

```python
college_num.info()
```

```python
list(college_cat)
```

```python
tempdf = pd.get_dummies(college_cat, columns=list(college_cat))
```

```python
tempdf
```

```python
from sklearn.preprocessing import StandardScaler

college_num_scaled = StandardScaler().fit_transform(college_num)
temp_df_num = pd.DataFrame(college_num_scaled, columns = list(college_num))
temp_df_num.head()
```

```python
# Place the DataFrames side by side
horizontal_stack = pd.concat([temp_df_num, tempdf], axis=1)
```

```python
horizontal_stack
```

```python
horizontal_stack.info()
```

```python
# split a dataset into train and test sets
from sklearn.model_selection import train_test_split


# split into train test sets
X_train, X_test, y_train, y_test = train_test_split(horizontal_stack, college_labels, test_size=0.33)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.linear_model import SGDClassifier

sgd_clf = SGDClassifier(max_iter=1000, tol=1e-3, random_state=42)
sgd_clf.fit(X_train, y_train)
```

```python
sgd_clf.predict(X_train)
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

cv_sgd = KFold(n_splits=10, shuffle=True, random_state=42)
cross_val_score(sgd_clf, X_train, y_train, cv=cv_sgd, scoring="accuracy", n_jobs=-1)
```

```python
scores = cross_val_score(sgd_clf, X_train, y_train, cv=cv_sgd, scoring="accuracy", n_jobs=-1)
print('Accuracy: %.3f (%.3f)' % (np.mean(scores), np.std(scores)))
```

```python
from sklearn.model_selection import KFold
from sklearn.tree import DecisionTreeClassifier

dec_tree_clf = DecisionTreeClassifier(random_state=42)

cv_dec_tree = KFold(n_splits=10, shuffle=True, random_state=42)
cross_val_score(dec_tree_clf, X_train, y_train, cv=cv_dec_tree, scoring="accuracy", n_jobs=-1)
```

```python
scores = cross_val_score(dec_tree_clf, X_train, y_train, cv=cv_dec_tree, scoring="accuracy", n_jobs=-1)
print('Accuracy: %.3f (%.3f)' % (np.mean(scores), np.std(scores)))
```

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression()

cv_log_reg = KFold(n_splits=10, shuffle=True, random_state=42)
cross_val_score(log_reg, X_train, y_train, cv=cv_log_reg, scoring="accuracy", n_jobs=-1)
```

```python
scores = cross_val_score(log_reg, X_train, y_train, cv=cv_log_reg, scoring="accuracy", n_jobs=-1)
print('Accuracy: %.3f (%.3f)' % (np.mean(scores), np.std(scores)))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics

random_clf = RandomForestClassifier(n_estimators=100)

random_clf.fit(X_train, y_train)

y_pred = random_clf.predict(X_test)

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
random_clf = RandomForestClassifier(n_estimators=100)

cv_rand_for = KFold(n_splits=10, shuffle=True, random_state=42)
cross_val_score(random_clf, X_train, y_train, cv=cv_rand_for, scoring="accuracy", n_jobs=-1)
```

```python
scores = cross_val_score(random_clf, X_train, y_train, cv=cv_rand_for, scoring="accuracy", n_jobs=-1)
print('Accuracy: %.3f (%.3f)' % (np.mean(scores), np.std(scores)))
```

```python
from sklearn.naive_bayes import GaussianNB

gaussian_nb_clf = GaussianNB()

cv_gaus_nb = KFold(n_splits=10, shuffle=True, random_state=42)
cross_val_score(gaussian_nb_clf, X_train, y_train, cv=cv_gaus_nb, scoring="accuracy", n_jobs=-1)
```
